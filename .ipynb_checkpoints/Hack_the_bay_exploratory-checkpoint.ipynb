```python
import pandas as pd
```

```python
water_df = pd.read_csv('../SourceMaterial/Water_FINAL.csv')
```

```python
column_names = []
for column in water_df.columns:
    column_names.append(column)
column_names
```

```python
water_temps = water_df[['Date', 'ModifiedDate', 'HUCNAME_', 'Latitude', 'Longitude', 'Parameter', "ParameterName_CBP", 'ParameterName_CMC', 'MeasureValue']].loc[water_df['Parameter'].str.contains('WT')]
water_temps
```

```python
ph_df = water_df[['Date', 'ModifiedDate', 'HUCNAME_', 'Latitude', 'Longitude', 'Parameter',"ParameterName_CBP", 'ParameterName_CMC', 'MeasureValue']].loc[water_df['Parameter'].str.contains("PH")]
ph_df.head()
```

```python
conductivity_df = water_df[['Date', 'ModifiedDate', 'HUCNAME_', 'Latitude', 'Longitude', 'Parameter',"ParameterName_CBP", 'ParameterName_CMC', 'MeasureValue']].loc[water_df['Parameter'].str.contains('CO')]
```

```python
conductivity_df.head()
```

```python
water_temps.to_csv('../SourceMaterial/water_temps.csv', index=False)
ph_df.to_csv('../SourceMaterial/ph.csv', index=False)
conductivity_df.to_csv('../SourceMaterial/conductivity.csv', index=False)
```
